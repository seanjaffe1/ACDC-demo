Visualise raw NIfTI volumes and segmentation masks.

```python
import sys
from pathlib import Path

# Resolve project root relative to this notebook's location
NOTEBOOK_DIR = Path('/Users/seanjaffe/Documents/ACDC-demo/notebooks')
ROOT = NOTEBOOK_DIR.parent
sys.path.insert(0, str(ROOT))

import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from src.utils.acdc_loader import ACDCDataset, PATHOLOGY_CLASSES

dataset = ACDCDataset(ROOT / 'data/raw')
print(f'Total subjects: {len(dataset)}')
for cls in PATHOLOGY_CLASSES:
    n = len(dataset.by_pathology(cls))
    print(f'  {cls}: {n}')
```

```python
# Visualise one subject — mid-slice at ED
subject = dataset[0]
(img, spacing), (seg, _) = subject.load_ed()
mid = img.shape[2] // 2

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(img[:, :, mid].T, cmap='gray', origin='lower')
axes[0].set_title(f'{subject.subject_id} — image (ED, z={mid})')
axes[1].imshow(seg[:, :, mid].T, cmap='tab10', vmin=0, vmax=3, origin='lower')
axes[1].set_title('Segmentation (1=RV, 2=Myo, 3=LV)')
plt.tight_layout()
```
